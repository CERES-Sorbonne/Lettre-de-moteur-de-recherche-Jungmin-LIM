```python
import pandas
import datetime
from datetime import date
import pandas as pd

date_str = date.today().strftime("%d/%m/%Y")
path_xls = "" #convertir excel file ver xls et stocker des informations

xls = pandas.ExcelFile(path_xls)
df = pandas.read_excel(xls, sheet_name ='Worksheet')

dic ={}

for line in df.iterrows():
    
    id_base = line[1][0]
    ID = str(id_base)
    value = list(line[1].values[1:])
    nom_infos = ["Surname","First name","Orignal_Location","Date_year","Month","Day","Index_page","Current_location","City","County","Unique_Identifier","Notes_written_on_the_letter","My_notes"]
    try:
        value[3] = int(value[3])
    except:
        pass
    try:
        value[4] = int(value[4])
    except:
        pass
    try:
        value[5] = int(value[5])
    except:
        pass
    try:
        value[6] = int(value[6])
    except:
        pass
    value = dict(zip(nom_infos, value))
    dic[ID] = value
as_list = [[cle, val["Surname"]] for cle, val in dic.items()] #stocker ID et vaule dans le dictionnaire

#importer json pour stocker le dictionnaire

import json
import re
for nom, struct in [["liste", as_list],["dico", dic]]:
    path_out = f"/{nom}_titires_ID.json"
    print(path_out, "w")
    w= open(path_out, "w")
    w.write(json.dumps(struct, indent = 2, ensure_ascii=False))
    w.close()
    
#apporter html et dictionnaire
with open(" myTableinitale1.html", encoding="utf-8") as f:
    header =f.read()
with open("/dico_titires_ID.json") as f:
    dic_titres = json.load(f)
out = ""
table_header = []
cpt = 0
test = []
for ID, infos in dic_titres.items():
    if len(table_header) == 0:
        table_header = infos.keys()
        out+=" <tr><th>ID</th><th>%s</th><tr>\n"%"</th><th>".join(table_header)#pour utiliser mytable, il faut mettre <tr>,<td> pertinent
    elems =[infos[cle] for cle in table_header]
    test.append([len(str(infos["Unique_Identifier"])),infos["Unique_Identifier"]])
    elems = [str(x) if str(x) not in ["null", "None", "nan"] else "" for x in elems]
    link = f"id=\"{ID}\""
    this_info = "</td><td>".join(elems)
    l = f" <tr><td>{ID}</td><td>{this_info}</td></tr>\n"
    out +=l

out_html = re.sub("{{content_table}}", out, header)
out_html = re.sub("{{date}}", date_str, out_html)
with open("testwithmytable.html", "w", encoding="utf-8") as w:
    w.write(out_html)
    

print(sorted(test) [-10:])

#référence Antonomaz "https://antonomaz.huma-num.fr"
```
